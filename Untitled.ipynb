```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as web
import datetime
```

```python
assets=['AMZN', 'ATR', 'BAC', 'BMY', 'CERN', 'CHRW', 'ADI', 'AMP', 'APA', 'BA', 'BAC', 'BHC', 'BLK', 'AA', 'AAL', 'AAPL', 'ABBV', 'ABT', 'ADBE', 'ADM', 'ADP', 'ADSK', 'AES', 'AFL', 'AIG'] #CQNS portfolio

q_assets=len(assets)

print('The total number of assets are {0}'.format(q_assets))

initial_date='2018-01-01'
present_date=datetime.date.today().strftime('%Y-%m-%d')

dataframe=pd.DataFrame()
for i in assets:
    dataframe[i] = web.DataReader(i,data_source='yahoo',start=initial_date , end=present_date)['Adj Close']
```

```python
dataframe
```

```python
df=dataframe.copy()
```

```python
plt.figure(figsize=(12,4)) 
for j in dataframe.columns.values:
    plt.plot( dataframe[j],  label=j)
plt.legend(dataframe.columns.values)
plt.xlabel('Period',fontsize=15)
plt.ylabel('Portfolio History',fontsize=15)
title='Portfolio optimization'
plt.title(title, fontsize=19)
```

```python
log_return = np.log(df/df.shift(1))       #log_return
log_return.head()
```

```python
meanReturns = log_return.mean() 
covMatrix = log_return.cov()
rfr = 0.011
```

```python
num_portfolios = 25000
all_weights = np.zeros((num_portfolios, len(dataframe.columns)))
ret_arr = np.zeros(num_portfolios)
vol_arr = np.zeros(num_portfolios)
sharpe_arr = np.zeros(num_portfolios)
weights = np.ones(q_assets-1)/(q_assets-1)

for ind in range(num_portfolios): 
    
    # expected return 
    ret_arr[ind] = np.sum((log_return.mean()*weights)*252)

    # expected volatility 
    vol_arr[ind] = np.sqrt(np.dot(weights.T,np.dot(log_return.cov()*252, weights)))

    # Sharpe Ratio 
    sharpe_arr[ind] = (ret_arr[ind]-rfr)/vol_arr[ind]
```

```python
plt.figure()
plt.scatter(vol_arr,ret_arr,c=sharpe_arr,marker='o')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
```
